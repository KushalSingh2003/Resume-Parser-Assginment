```python
def extract_email(text):
    email_pattern = re.compile(
        r'[a-zA-Z0-9._%+-]+@[a-zA-Z0-9.-]+\.(?:com|org|net|edu|gov|in|co\.in)'
    )

    matches = email_pattern.findall(text)
    return matches[0] if matches else ""
```

```python
import re
import spacy

nlp = spacy.load("en_core_web_sm")

name_pattern = re.compile(
    r"\b(?!Curriculum Vitae|Resume|RESUME|C.V.|Name|Full Name)[A-Z][a-zA-Z]+(?:\s[A-Z][a-zA-Z]+){0,2}\b"
)

def clean_name(name):
    name = re.sub(r'\s+', ' ', name).strip()
    name = re.sub(r'\b(Name|Full Name|Email|E-mail|Phone|Contact)\b', '', name, flags=re.IGNORECASE).strip()
    return name

def extract_name(text):
    name_match = name_pattern.search(text)
    if name_match:
        return clean_name(name_match.group())

    doc = nlp(text)
    for ent in doc.ents:
        if ent.label_ == "PERSON" and len(ent.text.split()) <= 3:
            return ent.text.strip()

    return ""
```

```python
# EDU_KEYWORDS = [
#     "Bachelor",
#     "Master",
#     "B.Sc",
#     "M.Sc",
#     "PhD",
#     "B.E",
#     "M.E",
#     "B.Tech",
#     "M.Tech",
#     "MBA"
# ]

# def extract_education(text):
#     lines = text.split('\n')
#     education = []

#     for line in lines:
#         for word in EDU_KEYWORDS:
#             if word.lower() in line.lower():
#                 education.append(line.strip())
#                 break  # Avoid duplicate entries

#     return education
def get_education_section(text):
    lower_text = text.lower()
    # Look for EDUCATION or ACADEMIC
    start = lower_text.find("education")
    if start == -1: start = lower_text.find("academic")
    if start == -1: return ""

    end = -1
    # More comprehensive list of sections that follow education
    next_keywords = ["experience", "skills", "projects", "certifications", "work", "internship", "achievements", "technical"]
    for kw in next_keywords:
        idx = lower_text.find(kw, start + 10)
        if idx != -1:
            if end == -1 or idx < end: end = idx

    if end == -1: end = len(text)
    return text[start:end]

def extract_education(text):
    education_text = get_education_section(text)
    education = []
    lines = [line.strip() for line in education_text.split('\n') if line.strip()]

    current_inst = ""
    current_deg = ""

    inst_kw = ["University", "College", "Institute", "School", "Academy", "Vidhyalaya", "Vidyalaya", "Kendriya", "DAV", "Public", "Convent"]
    deg_kw = ["B.E", "B.Tech", "Bachelor", "M.Tech", "Master", "MCA", "MBA", "B.Sc", "10th", "12th", "Intermediate", "High School"]

    for line in lines:
        if line.lower() in ["education", "academic"]: continue

        is_inst = any(kw.lower() in line.lower() for kw in inst_kw)
        is_deg = any(kw.lower() in line.lower() for kw in deg_kw)

        if is_inst:
            if current_inst: # New institution found, save previous
                education.append({"degree": current_deg, "institution": current_inst})
                current_deg = ""
            current_inst = re.sub(r'\d{4}.*$', '', line).strip()
        elif is_deg:
            current_deg = line if not current_deg else current_deg + " " + line

    if current_inst or current_deg:
        education.append({"degree": current_deg, "institution": current_inst})
    return education



```

```python
# def extract_experience(text):

#     experience_keywords = ['experience', 'work', 'internship', 'employment']

#     exp_lines = []

#     lines = text.split('\n')

#     for line in lines:

#         for keyword in experience_keywords:

#             if keyword.lower() in line.lower():

#                 exp_lines.append(line.strip())

#     return exp_lines
def extract_experience(text):
    section_headers = ['experience', 'work experience', 'employment']
    end_headers = ['education', 'projects', 'skills', 'technical skills', 'publications', 'certifications']

    lines = text.split('\n')
    exp_lines = []
    in_experience = False

    for line in lines:
        stripped = line.strip()
        lower = stripped.lower()

        # Check if we're entering the Experience section
        if any(lower == kw for kw in section_headers):
            in_experience = True
            continue

        # Check if we've hit the next section — stop collecting
        if in_experience and any(lower == kw for kw in end_headers):
            break

        if in_experience and stripped:
            exp_lines.append(stripped)

    return exp_lines
```

```python
import fitz

def extract_pdf_links(pdf_path):

    links = []

    try:

        doc = fitz.open(pdf_path)

        for page in doc:

            for link in page.get_links():

                if "uri" in link:
                    links.append(link["uri"])

        doc.close()

    except:
        pass

    return links
```

```python
import re

def extract_social_links(filepath, text):

    linkedin = ""
    github = ""

    # Regex to extract username from any github.com URL
    github_username_extractor_pattern = re.compile(
        r'(?:https?://)?(?:www\.)?github\.com/([a-zA-Z0-9_-]+)',
        re.IGNORECASE
    )

    # 1. Check hidden hyperlinks first (from PDF)
    links = extract_pdf_links(filepath)
    for link in links:
        if "linkedin.com" in link.lower() and not linkedin:
            linkedin = link

        # If a GitHub link is found in PDF, try to extract the username from it
        if "github.com" in link.lower() and not github:
            match = github_username_extractor_pattern.search(link)
            if match:
                username = match.group(1)
                github = f"https://github.com/{username}"
                # We prioritize the first username found in PDF links for the profile
                break

    # 2. Fallback to regex on text if GitHub profile not found in PDF links
    if not github:
        match = github_username_extractor_pattern.search(text)
        if match:
            username = match.group(1)
            github = f"https://github.com/{username}"

    # Fallback to regex for LinkedIn if not found in PDF links
    if not linkedin:
        linkedin_pattern = re.compile(
            r'https?://(?:www\.)?linkedin\.com/[^\s]+',
            re.IGNORECASE
        )
        linkedin_match = linkedin_pattern.search(text)
        if linkedin_match:
            linkedin = linkedin_match.group(0)

    return linkedin, github
```

```python
linkedin_profile, github_profile = extract_social_links('/content/Kush_Recent_Resume.pdf', cleaned)

parsed_resume = {

    'Name': extract_name(cleaned),

    'Email': extract_email(raw_text),

    'Phone': extract_phone(cleaned),

    'Skills': extract_skills(cleaned),

    'Education': extract_education(cleaned),

    'Experience': extract_experience(cleaned),

    'LinkedIn': linkedin_profile,

    'GitHub': github_profile

}

print(parsed_resume)
```
